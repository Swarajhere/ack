```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.flipkart.com/vivo-x200-pro-5g-titanium-grey-512-gb/p/itm1e8383eef06a4?pid=MOBH77FZC3YFRRHZ&lid=LSTMOBH77FZC3YFRRHZC3HYUQ&marketplace=FLIPKART&q=vivo+x200+pro&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&fm=organic&iid=7c7314ae-a94f-41b4-b983-13571bc62844.MOBH77FZC3YFRRHZ.SEARCH&ppt=clp&ppn=mobile-phones-store&ssid=aacdshquo00000001746550824998&qH=31c516fd626e2367')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

```python
title = soup.title.text
print(title)
```

```python
#reviews of the product
reviews = soup.findAll('div',{'class' : 'ZmyHeo'});
for review in reviews:
    print(review.get_text() + "\n\n")
    
```

```python
#average overall rating
ratings = soup.find('div',{'class' : 'col-12-12 ggs1+C'});
print(ratings.get_text())
```

```python
# individual ratings
individual_ratings = soup.findAll('div',{'class' : 'XQDdHH Ga3i8K'});
for individual_rating in individual_ratings:
    print(individual_rating.get_text() + "\n")
```

```python
# comment tags
comments = soup.findAll('p',{'class' : 'z9E0IG'});
for comment in comments:
    print(comment.get_text() + "\n")
```

```python
#customer name
customer_names = soup.find('p', {'class' : '_2NsDsF AwS1CA'});
print(customer_names.get_text())
```

```python
#q&a
questions = soup.findAll('div', {'class':'wys2hv _43gOsC'});
for question in questions:
    print(question.get_text())
    answer = question.findNext('div',{'class':'JxAXcP'});
    print(answer.get_text() + "\n")
```

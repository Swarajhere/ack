```python
import pandas as pd

# Load dataset with semicolon as delimiter
df = pd.read_csv("dataset_Facebook.csv", delimiter=';')

# Show first few rows
df.head()
```

```python
df.shape
```

### a) Create Data Subsets for Type of Post

```python
# Get unique post types
print(df['Type'].unique())
```

```python
# Create subsets
photo_posts = df[df['Type'] == 'Photo']
status_posts = df[df['Type'] == 'Status']
video_posts = df[df['Type'] == 'Video']
link_posts = df[df['Type'] == 'Link']
```

```python
print(link_posts.shape)
link_posts.head()
```

```python
# Save subsets to verify
photo_posts.to_csv('photo_posts.csv', index=False)
```

### (b) Merge Two Subsets (e.g., Photo and Status)

```python
merged_df = pd.concat([photo_posts, status_posts])
merged_df.head()
```

### (c) Sort Data on Page Total Likes

```python
sorted_df = df.sort_values(by='Page total likes', ascending=False)
sorted_df.head()
```

### (d) Transposing Data

```python
# Transposing Data
transposed_df = df.transpose()
transposed_df.head()
```

### (e) Melting Data to Long Format

```python
melted_df = pd.melt(
    df,
    id_vars=['Type', 'Category', 'Post Month'],
    value_vars=['Lifetime Post Total Reach', 'Lifetime Post Total Impressions', 'Total Interactions'],
    var_name='Metric',
    value_name='Value'
)

# Save melted data to verify
# melted_df.to_csv('melted_data.csv', index=False)
```

```python
melted_df.head()
```

### (f) Casting Data to Wide Format

```python
wide_df = melted_df.pivot_table(
    index=['Type', 'Category', 'Post Month'],
    columns='Metric',
    values='Value',
    aggfunc='mean'
).reset_index()

wide_df.head()
```

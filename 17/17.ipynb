```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("heart.csv")
```

```python
# a. Histogram
plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Patients")  # Objective: Show age distribution to assess range and skewness
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# b. Dot Plot
plt.figure(figsize=(6,4))
sns.stripplot(x=df['target'], y=df['chol'], color='purple')  # Changed x to target, y to chol
plt.title("Cholesterol by Heart Disease")  # Objective: Compare cholesterol levels by disease status
plt.xlabel("Heart Disease (0=No, 1=Yes)")
plt.ylabel("Cholesterol (chol)")
plt.show()
```

```python
# c. Bar Plot
# Objective: Compare average cholesterol by sex (0=female, 1=male).

df.groupby('sex')['chol'].mean().plot(kind='bar', color=['pink', 'blue'])
plt.title("Average Cholesterol by Sex")
plt.ylabel("Cholesterol")
plt.xticks([0,1], ['Female', 'Male'], rotation=0)
plt.show()
```

```python
# d. Line Chart
# Objective: Show trend of max heart rate with age
plt.figure(figsize=(6,4))
age_thalach = df.groupby('age')['thalach'].mean()
plt.plot(age_thalach.index, age_thalach, color='red', label='Max Heart Rate')  # Changed x to age, y to mean thalach
plt.title("Average Max Heart Rate by Age")  
plt.xlabel("Age")
plt.ylabel("Average Max HR (thalach)")
plt.legend()
plt.show()
```

```python
# e. Box Plot with Histogram and Scatter Plot
# Objective: Visualize the distribution and outliers in cholesterol levels (chol) grouped by heart disease (target) using box plots, histograms, and scatter points.
```

```python
plt.figure(figsize=(8, 6))

# Boxplot for cholesterol by heart disease status
sns.boxplot(x='target', y='chol', data=df, palette='pastel')

# Scatter plot on top of boxplot (jittered)
sns.stripplot(x='target', y='chol', data=df, color='gray', size=4, jitter=True)

plt.title("Cholesterol Levels by Heart Disease Status")
plt.xlabel("Heart Disease (0=No, 1=Yes)")
plt.ylabel("Cholesterol (chol)")
plt.show()

# Optional: Histogram separately to show full distribution
plt.figure(figsize=(6, 4))
sns.histplot(df['chol'], bins=20, kde=True, color='orange')
plt.title("Cholesterol Distribution of All Patients")
plt.xlabel("Cholesterol (chol)")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("adult_dataset.csv")
```

```python
# f. Histogram
# Objective: Visualize the distribution of age in the population
plt.figure()
df['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# g. Dot Plots
plt.figure(figsize=(6, 4))
# Subsample to 1000 rows for less congestion
df_sample = df.dropna().sample(n=1000, random_state=42)
sns.swarmplot(x='income', y='hours-per-week', data=df_sample, color='blue', alpha=0.5, size=4)
plt.title("Dot Plot of Hours per Week by Income")
plt.xlabel("Income")
plt.ylabel("Hours per Week")
plt.show()
# Objective: Visualize hours worked per week across income levels to compare work intensity.
```

```python
# h. Bar Plot
# Objective: Compare the count of income groups
plt.figure()
sns.countplot(x='income', hue='income', data=df, palette='Set2', legend=False)
plt.title("Bar Plot of Income Categories")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
```

```python
# i. Line Charts
plt.figure(figsize=(6, 4))
age_hours = df.groupby('age')['hours-per-week'].mean().dropna()
plt.plot(age_hours.index, age_hours.values, color='purple')
plt.title("Line Chart of Average Hours per Week by Age")
plt.xlabel("Age")
plt.ylabel("Average Hours per Week")
plt.show()
# Objective: Show the trend of average hours worked per week across ages to identify work patterns.
```

```python
# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Box plot with swarm plot overlay
sns.boxplot(x='income', y='hours-per-week', data=df, ax=ax1)
df_sample = df.sample(n=500, random_state=42)  # Subsample for swarm
sns.swarmplot(x='income', y='hours-per-week', data=df_sample, color='blue', size=3, alpha=0.5, ax=ax1)
ax1.set_title("Box & Scatter: Hours per Week by Income")
ax1.set_xlabel("Income")
ax1.set_ylabel("Hours per Week")

# Histogram for each income group
sns.histplot(data=df[df['income'] == '<=50K'], x='hours-per-week', bins=20, color='skyblue', ax=ax2, label='<=50K')
sns.histplot(data=df[df['income'] == '>50K'], x='hours-per-week', bins=20, color='salmon', ax=ax2, label='>50K', alpha=0.5)
ax2.set_title("Histogram: Hours per Week by Income")
ax2.set_xlabel("Hours per Week")
ax2.set_ylabel("Frequency")
ax2.legend()

plt.tight_layout()
plt.show()
```

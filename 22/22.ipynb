```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("adult_dataset.csv")

# Basic cleaning
df = df.replace('?', pd.NA).dropna()
```

```python
# a. Histograms
# Objective: Show the distribution of ages in the dataset.
plt.figure()
df['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# b. Pie Chart
# Objective: Show proportion of income categories.
plt.figure()
df['income'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title("Pie Chart of Income Distribution")
plt.ylabel("")
plt.show()
```

```python
# c. Box Plots
plt.figure(figsize=(6, 4))
sns.boxplot(x='income', y='hours-per-week', data=df.dropna(), hue='income', palette='Set3')
plt.title("Box Plot of Hours per Week by Income")
plt.xlabel("Income")
plt.ylabel("Hours per Week")
plt.show()
# Objective: Compare the distribution of hours worked per week across income levels to identify differences in work intensity.
```

```python
# d. Scatter Plot
# Objective: Analyze relationship between age and hours-per-week.

plt.figure()
sns.scatterplot(x='age', y='hours-per-week', data=df, alpha=0.5)
plt.title("Scatter Plot of Age vs Hours per Week")
plt.xlabel("Age")
plt.ylabel("Hours per Week")
plt.show()
```

```python
# d. Scatter Plots
plt.figure(figsize=(6, 4))
df_sample = df.dropna().sample(n=1000, random_state=42)  # Subsample for clarity
plt.scatter(df_sample['age'], df_sample['hours-per-week'], color='blue', alpha=0.5)
plt.title("Scatter Plot of Age vs. Hours per Week")
plt.xlabel("Age")
plt.ylabel("Hours per Week")
plt.show()
# Objective: Show the relationship between age and hours worked per week to explore work patterns.
```

```python
# e. Add Boxplots to a Scatterplot
g = sns.JointGrid(data=df_sample, x='age', y='hours-per-week', height=6)
g.plot_joint(sns.scatterplot, color='blue', alpha=0.5)
g.ax_marg_x.boxplot(df_sample['age'], vert=False, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightblue', edgecolor='black'))
g.ax_marg_y.boxplot(df_sample['hours-per-week'], vert=True, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightblue', edgecolor='black'))
g.ax_marg_x.set_yticks([])  # Hide y-ticks on top margin
g.ax_marg_y.set_xticks([])  # Hide x-ticks on right margin
g.set_axis_labels("Age", "Hours per Week")
plt.suptitle("Scatterplot of Age vs. Hours per Week with Boxplot Margins")  # Objective: Show the relationship between age and hours per week with marginal boxplots for their distributions
plt.show()
```

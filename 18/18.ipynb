```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("heart.csv")
```

```python
# a. Histogram
plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Patients")  # Objective: Show age distribution to assess range and skewness
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# b. Pie Chart
# Objective: Show gender ratio in dataset.

df['sex'].value_counts().plot.pie(labels=['Male', 'Female'], autopct='%1.1f%%', colors=['lightblue','pink'])
plt.title("Gender Distribution")
plt.ylabel('')
plt.show()
```

```python
# c. Box Plot
# Objective: Show spread of cholesterol values.

sns.boxplot(y=df['chol'], color='lightblue')
plt.title("Box Plot of Cholesterol")
plt.show()
```

```python
# d. Scatter Plot
# Objective: Show relation between age and cholesterol.

plt.scatter(df['age'], df['chol'], alpha=0.6, color='teal')
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

```python
# e. Add Boxplots to Scatterplot
g = sns.JointGrid(data=df, x='age', y='chol', height=6)
# Scatterplot
g.plot_joint(sns.scatterplot, color='blue')
# Boxplot for age (top margin)
g.ax_marg_x.boxplot(df['age'], vert=False, widths=0.2, patch_artist=True, 
                    boxprops=dict(facecolor='lightblue', edgecolor='black'))
g.ax_marg_x.set_yticks([])  # Hide y-ticks on top margin
# Boxplot for chol (right margin)
g.ax_marg_y.boxplot(df['chol'], vert=True, widths=0.2, patch_artist=True, 
                    boxprops=dict(facecolor='lightblue', edgecolor='black'))
g.ax_marg_y.set_xticks([])  # Hide x-ticks on right margin
g.set_axis_labels("Age", "Cholesterol (chol)")
plt.suptitle("Scatterplot of Age vs. Cholesterol with Boxplot Margins")  # Objective: Show the relationship between age and cholesterol with marginal boxplots for their distributions
plt.show()
```

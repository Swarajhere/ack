```python
import pandas as pd

df = pd.read_csv('adult_dataset.csv')

df.head()
```

```python
df.shape
```

```python
us_df = df[df['native-country'] == 'United-States']
india_df = df[df['native-country'] == 'India'] 

male_df = df[df['gender'] == 'Male']
female_df = df[df['gender'] == 'Female']

white_df = df[df['race'] == 'White']
black_df = df[df['race'] == 'Black']
```

```python
us_df.shape
```

```python
# Merge `us_df` and `male_df` (People from the US who are Male)
merged_df = pd.concat([us_df, male_df])

# View merged data shape
merged_df.shape
```

```python
# 12u. Sort Data using age
sorted_df = df.sort_values(by='age')

sorted_df.head()
```

```python
#  12v. Transpose Data
transposed_df = df.transpose()
transposed_df.head()
```

```python
melted_df = pd.melt(
    df,
    id_vars=['native-country', 'gender', 'race'],
    value_vars=['age', 'capital-gain', 'capital-loss', 'hours-per-week'],
    var_name='Feature',
    value_name='Value'
)

melted_df.head()
```

```python
wide_df = melted_df.pivot_table(
    index=['native-country', 'gender', 'race'],
    columns='Feature',
    values='Value',
    aggfunc='mean'
).reset_index()

wide_df.head()
```

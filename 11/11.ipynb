```python
import pandas as pd

df = pd.read_csv('movies_metadata.csv')

df.head()
```

```python
print(df['original_language'].unique())
```

```python
# m. Create data subsets for different languages
subset_en = df[df['original_language'] == 'en']
subset_fr = df[df['original_language'] == 'fr']
```

```python
subset_en.head()
```

```python
# n. Merge two subsets (e.g., English and French movies)
merged_df = pd.concat([subset_en, subset_fr])
```

```python
# o. Sort Data using customer ratings (vote_average)
sorted_df = df.sort_values(by='vote_average', ascending=False)
```

```python
sorted_df.head()
```

```python
# p. Transposing Data
transposed_df = df.transpose()
transposed_df.head()
```

```python
# q. Melting Data to long format (select a few columns)
melted_df = pd.melt(
    df,
    id_vars=['id', 'original_language'],
    value_vars=['budget', 'revenue', 'runtime', 'vote_average'],
    var_name='Feature',
    value_name='Value'
)

melted_df.head()
```

```python
# Convert the 'Value' column to numeric, coercing errors if necessary
melted_df['Value'] = pd.to_numeric(melted_df['Value'], errors='coerce')

# r. Casting data back to wide format
wide_df = melted_df.pivot_table(
    index=['id', 'original_language'],
    columns='Feature',
    values='Value'
).reset_index()

wide_df.head()
```

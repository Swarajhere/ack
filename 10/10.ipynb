```python
import pandas as pd

# # Define column names based on the data description
# column_names = [
#     'SepalLengthCm', 'SepalWidthCm', 
#     'PetalLengthCm', 'PetalWidthCm', 
#     'Species'
# ]

# Load the dataset without header row
df = pd.read_csv('iris1.csv')

# Add an ID column to uniquely identify rows
# df.insert(0, 'Id', range(1, 1 + len(df)))

# Display first few rows
df.head()
```

### 10g. Create data subsets for different species

```python
# Subset for Iris-setosa
setosa_df = df[df['variety'] == 'Setosa']

# Subset for Iris-versicolor
versicolor_df = df[df['variety'] == 'Versicolor']

# Subset for Iris-virginica
virginica_df = df[df['variety'] == 'Virginica']
```

```python
setosa_df.head()
```

### 10h. Merge two subsets

```python
# Merge Setosa and Versicolor
merged_df = pd.concat([setosa_df, versicolor_df])

# View merged data
merged_df.head()
```

### 10i. Sort Data by Petal Length

```python
# Sort by PetalLengthCm
sorted_df = df.sort_values(by='petal.length')

# View sorted data
sorted_df.head()
```

### 10j. Transpose Data

```python
# Transposing Data
transposed_df = df.transpose()
transposed_df.head()
```

### 10k. Melting Data to Long Format

```python
melted_df = pd.melt(
    df,
    id_vars=['variety'],
    value_vars=['sepal.length', 'sepal.width', 'petal.length', 'petal.width'],
    var_name='Measurement',
    value_name='Value'
)
# Save melted data to verify
# melted_df.to_csv('melted_data.csv', index=False)
melted_df.head()
```

### 10l. Casting Data to Wide Format

```python
# Pivot to wide format
wide_df = melted_df.pivot_table(
    index=['variety'],
    columns='Measurement',
    values='Value'
).reset_index()

# Save wide format if needed
#wide_df.to_csv('iris_wide.csv', index=False)

# Display
wide_df.head()
```
